To work with categorical variables, we need to convert it into numerical one using one-hot encoding method as it is shown below:

```python
def dummies(df,z):
    state=pd.get_dummies(df[z],drop_first=True)
    df=pd.concat([df,state],axis=1)
    df.drop([z],axis=1,inplace=True)
    return df

cars=dummies(cars,'fueltype')
cars=dummies(cars,'aspiration')
cars=dummies(cars,'carbody')
cars=dummies(cars,'drivewheel')
cars=dummies(cars,'enginelocation')
cars=dummies(cars,'enginetype')
cars=dummies(cars,'cylindernumber')
cars=dummies(cars,'fuelsystem')
```

```python
x1=cars.iloc[:,:7]
x2=cars.iloc[:,8:]
y=cars.iloc[:,7]
```

```python
x=pd.concat([x1,x2],axis=1)
```

To normalize the range of independent variables of features we can use feature scaling method - Standardization (Z-score).
<img src="standardization.png" style="width:250px;height:200px">

```python
scaled_x=x.copy()
col_names=['carlength','carwidth','curbweight','enginesize','horsepower','fuelefficiency(1/100km)']
features=scaled_x[col_names]
scaler=StandardScaler()
features=scaler.fit_transform(features)

scaled_x[col_names]=features
print(f"Peak to Peak range by column in Raw X:\n{np.ptp(x,axis=0)}")   
print(f"Peak to Peak range by column in Normalized X:\n{np.ptp(scaled_x,axis=0)}")
```

```python
gr_x_train, gr_x_test, gr_y_train, gr_y_test=train_test_split(scaled_x,y,test_size=0.2,shuffle=True,random_state=123)
## Using Stratified Splits for obtaining training data, testing data
```

Next, we stratch gradient descent algorithm for multiple linear regression.
<img src="gd1.png" style="width:350px;height:80px">
<img src="gd2.png" style="width:450px;height:120px">
<img src="gd3.png" style="width:350px;height:400px">

```python
def LR_with_gradient_descent(X,y,alpha,epoch):
    m = X.shape[0]  # number of samples
    ones =np.ones((m,1))  
    X = np.concatenate((ones, X), axis=1)   # Nuw X with X0's =1
    n = X.shape[1]   # 
    Theta = np.ones(n)    # n= 4  parameter initialization
    h = np.dot(X, Theta)   # Compute hypothesis

  # Gradient descent algorithm
    cost = np.ones(epoch)
    for i in range (0, epoch):
        Theta[0] = Theta[0] - (alpha / X.shape[0]) * sum(h-y)
        for j in range(1, n):
            Theta[j]= Theta[j] - (alpha/ X.shape[0]) * sum((h-y) * X[:, j])
        h  = np.dot(X, Theta)
        cost[i] = 1/(2*m) * sum(np.square(h-y))  # Compute Cost
    return cost, Theta
```

```python
cost, theta= LR_with_gradient_descent(gr_x_train, gr_y_train, 0.3,2000)
print(theta)
```

```python
plt.plot(cost)
plt.xlabel('Number of iterations')
plt.ylabel('Cost function')
plt.show()
print(np.min(cost))
```

```python
x_select=gr_x_test
```

```python
onon=np.ones((41,1))
x_select=np.concatenate((onon,x_select),axis=1)
x_select
```

```python
r2_score(gr_y_test,np.dot(x_select,theta))
```

91.6% accuracy;
